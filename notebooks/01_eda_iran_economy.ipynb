```python
plt.figure(figsize=(10,4))
plt.plot(df["Year"], df["GDP_Current_USD"])
plt.title("Iran GDP (Current USD)")
plt.xlabel("Year")
plt.ylabel("USD")
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df["Year"], df["Inflation_Rate_Percent"], color = "green")
plt.title("Inflation Rate Over Time")
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df["Year"], df["Official_Exchange_Rate"])
plt.title("Official Exchange Rate (IRR/USD)")
plt.show()
```
