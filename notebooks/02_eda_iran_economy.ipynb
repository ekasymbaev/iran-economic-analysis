##Economic Eras

```python
df["Decade"] = (df["Year"] // 10) * 10
df.tail()
```

```python
df.groupby("Decade")[["GDP_Growth_Percent", "Inflation_Rate_Percent"]].mean()
```

```python
sns.scatterplot(
    data=df,
    x="Inflation_Rate_Percent",
    y="GDP_Growth_Percent"
)
```

```python
df.columns
```

```python
df["Trade_Balance"] = df["Exports_USD"] - df["Imports_USD"]
```

```python
plt.figure(figsize=(12,5))
plt.plot(df["Year"], df["Trade_Balance"], linewidth=2)
plt.axhline(0, linestyle="--")
plt.title("Iran Trade Balance Over Time")
plt.grid(alpha=0.3)
plt.show()
```

```python
corr = df.drop(columns=["Year"]).corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Correlation Matrix")
plt.show()
```

```python
df["High_Inflation"] = df["Inflation_Rate_Percent"] > df["Inflation_Rate_Percent"].quantile(0.95)
```
